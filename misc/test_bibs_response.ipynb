```python
from typing import List, Optional, Dict, Union, Any
from pydantic import BaseModel, validator, Field

class MarcSubField(BaseModel):
    code: Optional[str] = None
    data: Optional[str] = None

class FieldData(BaseModel):
    subfields: List[MarcSubField] = []
    ind1: Optional[str] = None
    ind2: Optional[str] = None

class MarcField(BaseModel):
    tag: str
    data: Union[str, FieldData, Dict[str, Any]] = Field(...)

class Marc(BaseModel):
    leader: str
    fields: List[MarcField]

class Language(BaseModel):
    code: str
    name: Optional[str]

class MaterialType(BaseModel):
    code: str
    value: Optional[str]

class BibLevel(BaseModel):
    code: str
    value: Optional[str]

class Country(BaseModel):
    code: str
    name: str

class Location(BaseModel):
    code: str
    name: str

class OrderInfo(BaseModel):
    orderId: str
    location: Location
    copies: int
    date: Optional[str]

class FixedFieldVal(BaseModel):
    value: str

class FixedField(BaseModel):
    label: str
    value: Optional[FixedFieldVal]
    display: Optional[str]

class SubField(BaseModel):
    tag: str
    content: str

class VarField(BaseModel):
    fieldTag: str
    marcTag: Optional[str]
    ind1: Optional[str]
    ind2: Optional[str]
    content: Optional[str]
    subfields: Optional[List[SubField]]

class Bib(BaseModel):
    id: str
    updatedDate: Optional[str]
    createdDate: Optional[str]
    deletedDate: Optional[str]
    deleted: bool
    suppressed: Optional[bool]
    available: Optional[bool]
    isbn: Optional[str]
    issn: Optional[str]
    upc: Optional[str]
    lang: Optional[Language]
    title: Optional[str]
    author: Optional[str]
    marc: Optional[Marc]
    materialType: Optional[MaterialType]
    bibLevel: Optional[BibLevel]
    publishYear: Optional[int]
    catalogDate: Optional[str]
    country: Optional[Country]
    orders: Optional[List[OrderInfo]]
    normTitle: Optional[str]
    normAuthor: Optional[str]
    locations: Optional[List[Location]]
    holdCount: Optional[int]
    copies: Optional[int]
    callNumber: Optional[str]
    volumes: Optional[List[str]]
    items: Optional[List[str]]
    fixedFields: Optional[Dict[int, FixedField]]
    varFields: Optional[List[VarField]]

class BibResultSet(BaseModel):
    total: Optional[int]
    start: Optional[int]
    entries: List[Bib]
```
